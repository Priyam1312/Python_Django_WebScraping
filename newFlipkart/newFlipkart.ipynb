```python
from bs4 import BeautifulSoup
import requests
```

```python
with open ("newflipkart.html",encoding="utf-8") as f:
      data=f.read()
```

```python
soup=BeautifulSoup(data,"html.parser")
```

```python
z=soup.find_all("div",class_="tUxRFH")

name=[]
memory=[]
color=[]
star=[]
review=[]
old=[]
new=[]
dis=[]
exc=[]
for i in z:
    n=i.find("div",class_="KzDlHZ")
    name.append(n.text.split("(")[0])
    m=i.find("li",class_="J+igdf").text[:10]
    memory.append(m)
    c1=i.find("div",class_="KzDlHZ").text.split("(")[1]
    c2=c1.split(",")[0]
    color.append(c2)
    s=i.find("div",class_="XQDdHH").text
    star.append(s)
    r=i.find("span",class_="Wphh3N").text.strip().split("&")[1]
    review.append(r.split(" ")[0])
    o=i.find("div",class_="yRaY8j ZYYwLA").text.split("₹")[1]
    old.append(int(o.replace(",","")))
    n=i.find("div",class_="Nx9bqj _4b5DiR").text.split("₹")[1]
    new.append(int(n.replace(",","")))
    d=i.find("div",class_="UkUFwK").text.strip().split("%")[0]
    dis.append(d)
    e=i.find_all("div",class_="yiggsN O5Fpg8")[-2].text.replace("Saver Deal","Nan")
    e1=e.replace("₹","").replace(",","")
    exc.append(e1)

import pandas as pd
import numpy as np
```

```python
exc
```

```python
data={"name":name,"memory":memory,"color":color,"star":star,"review":review,"old":old,"new":new,"dis":dis,"exc":exc}
```

```python
df=pd.DataFrame(data)
```

```python
df["diff"]=df["old"]-df["new"]
```
