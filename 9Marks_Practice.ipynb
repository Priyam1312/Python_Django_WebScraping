```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
file=open("FLIPKART_TV.html",encoding="utf-8")
response=file.read()
soup=BeautifulSoup(response,"html.parser")
a=soup.find_all("a",class_="CGtC98")
name=[]
size=[]
stars=[]
reviews=[]
actual_price=[]
final_price=[]
launch_year=[]
discount=[]
for i in a:
    name.append(i.find("div",class_="KzDlHZ").text.strip().split("(")[0].strip())
    size.append(i.find("div",class_="KzDlHZ").text.strip().split("(")[1].split(")")[0])
    try:
        stars.append(i.find("div",class_="XQDdHH").text.strip())
    except:
        stars.append(np.nan)
    try:
        reviews.append(i.find("span",class_="Wphh3N").text.strip().split()[3])
    except:
        reviews.append(np.nan)
    actual_price.append(i.find("div",class_="yRaY8j ZYYwLA").text.strip()[1:])
    final_price.append(i.find("div",class_="Nx9bqj _4b5DiR").text.strip()[1:])
    launch_year.append(i.find("li",class_="J+igdf").text.strip().split("Launch Year")[1][2:6])
    discount.append(i.find("div",class_="UkUFwK").text.strip().split()[0])

df=pd.DataFrame({"Name":name,"Size":size,"Stars":stars,"Reviews":reviews,"Actual Price":actual_price,"Final Price":final_price,"Launch Year":launch_year,"Discount":discount})
df
```
