```python
import requests
from bs4 import BeautifulSoup
response=requests.get("https://subslikescript.com/")
response=response.text
movies=[]
links=[]
soup=BeautifulSoup(response,"html.parser")
m=soup.find("ul",class_="scripts-list")
l=m.find_all("a",href=True)
for i in l:
    movies.append(i.text.strip())
    links.append("https://subslikescript.com/"+i["href"])
import pandas as pd
df=pd.DataFrame({"Movies":movies,"Links":links})
df
```

```python
import requests
from bs4 import BeautifulSoup
response=requests.get("https://www.politifact.com/")
response=response.text
soup=BeautifulSoup(response,"html.parser")
```

```python
source=[]
statement=[]
date=[]
link=[]
a=soup.find("div",class_="embla")
b=a.find_all("div",class_="embla__slide caro-listing__item")
for i in b:
    s=i.find("div",class_="m-statement__meta").find("a").text.strip()
    source.append(s)
    statement.append(i.find("div",class_="m-statement__quote").find("a").text.strip())
    link.append("https://www.politifact.com/"+i.find("div",class_="m-statement__quote").find("a")["href"])
    d=i.find("div",class_="m-statement__desc").text.strip()
    l=d.split()
    date.append(" ".join(l[2:5]))
import pandas as pd
df=pd.DataFrame({"Source":source,"Statement":statement,"Link":link,"Date":date})
print(df)
```

```python
from bs4 import BeautifulSoup
file=open("30 Most Popular Movies Right Now_ What to Watch In Theaters and Streaming __ Rotten Tomatoes – Movie and TV News.html",encoding="utf-8")
response=file.read()
soup=BeautifulSoup(response,"html.parser")
movie=[]
rating=[]
director=[]
a=soup.find_all("div",class_="col-sm-18 col-full-xs countdown-item-content")
for i in a:
    movie.append(i.find("h2").find("a").text.strip())
    rating.append(i.find("span",class_="tMeterScore").text.strip())
    director.append(i.find("div",class_="info director").find("a").text.strip())

import pandas as pd
df=pd.DataFrame({"Movies":movie,"Rating":rating,"Director Name":director})
df
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
file=open("Certified used Mercedes-Benz for sale.html",encoding="utf-8")
response=file.read()
soup=BeautifulSoup(response,"html.parser")
carname=[]
reviewcount=[]
rating=[]
price=[]
dealername=[]
mileage=[]
a=soup.find_all("div",class_="vehicle-details")
for i in a:
    carname.append(i.find("h2",class_="title").text.strip())
    mileage.append(i.find("div",class_="mileage").text.strip())
    reviewcount.append(i.find("span",class_="sds-rating__link sds-button-link").text.split(" ")[0][1:])
    dealername.append(i.find("div",class_="dealer-name").text.strip())
    price.append(i.find("span",class_="primary-price").text.strip())
    try:
        rating.append(i.find("span",class_="sds-rating__count").text.strip())
    except:
        rating.append(np.nan)
        
df=pd.DataFrame({"CarName":carname,"Mileage":mileage,"ReviewCount":reviewcount,"DealerName":dealername,"Price":price,"Rating":rating})
df
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
file=open("SENSEX Top Gainers, BSE Daily Top Gainers - Stock Market (1).html",encoding="utf-8")
response=file.read()
soup=BeautifulSoup(response,"html.parser")
a=soup.find("table")
b=a.find_all("tr")
company=[]
price=[]
prevclose_date=[]
prevclose_price=[]
change=[]
changeper=[]
daysh=[]
daysl=[]
for i in b[1:]:
    company.append(i.find("td",class_="v-align").text.strip())
    price.append(i.find("td",class_="txt-right v-align").text.strip())
    prevclose_date.append(i.find("td",class_="txt-right no-wrap").text.strip().split("\n")[1])
    prevclose_price.append(i.find("td",class_="txt-right no-wrap").text.strip().split("\n")[0])
    daysl.append(i.find("span",class_="fl f-9").text.strip())
    daysh.append(i.find("span",class_="fr f-9").text.strip())
    change.append(i.find("span",class_="green-btn").text.strip())
    c=i.find_all("td",class_="txt-ctr")
    for j in c:
        x=j.text.strip()
    changeper.append(x)
df=pd.DataFrame({"Company":company,"Price":price,"PrevCloseDate":prevclose_date,"PrevClosePrice":prevclose_price,"Change":change,"ChangePer":changeper,"DaysHigh":daysh,"DaysLow":daysl})
df
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
file=open("flipkart.html",encoding="utf-8")
response=file.read()
soup=BeautifulSoup(response,"html.parser")
a=soup.find_all("div",class_="_3pLy-c row")
phonename=[]
flipprice=[]
actualprice=[]
diff=[]
for i in a:
    phonename.append(i.find("div",class_="_4rR01T").text.strip())
    flipprice.append(i.find("div",class_="_30jeq3 _1_WHN1").text.strip())
    try:
        actualprice.append(i.find("div",class_="_3I9_wc _27UcVY").text.strip())
    except:
        actualprice.append(np.nan)
df=pd.DataFrame({"Name":phonename,"FP":flipprice,"AP":actualprice})
df=df.fillna(0)
df["FP"]=(df["FP"]
        .astype(str)
        .str.replace("₹","",regex=False)
        .str.replace(",","",regex=False))
df["FP"]=pd.to_numeric(df["FP"],errors="coerce")
df["AP"]=(df["AP"]
        .astype(str)
        .str.replace("₹","",regex=False)
        .str.replace(",","",regex=False))
df["AP"]=pd.to_numeric(df["AP"],errors="coerce")
df["Diff"]=df["AP"]-df["FP"]
df
```
